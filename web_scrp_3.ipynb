```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_='wikitable sortable plainrowheaders')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_title = table.find_all('th')
```

```python
world_table_titles = [title.text for title in world_title]

print(world_table_titles)
```

```python
world_table_titles = [title.text.strip() for title in world_title]

print(world_table_titles[0:10])
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(columns = world_table_titles[0:9])
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
i = 1
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    # print(len(individual_row_data))
    
    length = len(df)
    df.loc[length] = [i] + individual_row_data
    i = i+1
```
